```python
import google.generativeai as genai

genai.configure(api_key="your api key")

def two_agent_conversation():
    print("üí¨ TWO-AGENT CONVERSATION")
    print("=" * 40)

    # Create two agents
    agent1 = genai.GenerativeModel('gemini-2.0-flash')
    agent2 = genai.GenerativeModel('gemini-2.0-flash')

    topic = input("Enter conversation topic (or press enter for default): ").strip()
    if not topic:
        topic = "the future of technology"

    print(f"\nTopic: {topic}")
    print("Agent 1 and Agent 2 are having a conversation...\n")

    # Start conversation
    prompt1 = f"Start a conversation about {topic} as Agent 1. Give an opening statement:"
    response1 = agent1.generate_content(prompt1)
    print(f"Agent 1: {response1.text}")

    # Continue conversation for several turns
    turns = 4
    current_response = response1.text

    for turn in range(turns):
        # Agent 2 responds
        prompt2 = f"Agent 1 said: '{current_response}'. Respond as Agent 2 in a natural conversation about {topic}:"
        response2 = agent2.generate_content(prompt2)
        print(f"Agent 2: {response2.text}")

        # Agent 1 responds
        prompt1 = f"Agent 2 said: '{response2.text}'. Continue the conversation as Agent 1 about {topic}:"
        response1 = agent1.generate_content(prompt1)
        print(f"Agent 1: {response1.text}")

        current_response = response1.text

# Run simple conversation
two_agent_conversation()
```
